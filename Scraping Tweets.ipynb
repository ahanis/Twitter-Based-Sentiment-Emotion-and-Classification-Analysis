#Clean tweets

```python
def cleanTweet(tweet):

    #convert everything to lowercase
    tweet = tweet.lower() 
    
    #remove mentions   
    tweet = re.sub('@[\w]*','',tweet)  
    
    #remove URLs
    tweet = re.sub(r'https?:\/\/.*\/\w*', '', tweet)

    #remove numbers
    tweet = re.sub("\d+\s|\s\d+", " ", tweet)

    #remove hashtags
    tweet = re.sub('#\S+', '', tweet)

    #remove quotes
    tweet = tweet.replace('"', '')   

    #remove \n
    tweet = tweet.replace('\n', '') 

    #remove extra whitespace
    tweet = re.sub('\s+', ' ', tweet)
    
    #remove punctuations
    #tweet = re.sub(r"[,.;':@#?!\&/$]+\%*", ' ', tweet) 

    #remove punctuations
    tweet = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), '', tweet) 

    return tweet

#clean the tweets in the text column
df_modified['text'] = df_modified['text'].apply(cleanTweet)
```

```python
 def cleanTweet2(tweet):

    #replace emojis with text

    #list all possible emojis in a list
    all_emojis = list(UNICODE_EMOJI.keys())
    
    #count number of emojis in a tweet
    def emoji_count(tweet):
      emoji_counter = 0
      for word in tweet:
          if any(char in emoji.UNICODE_EMOJI for char in word):
              emoji_counter += 1
      return emoji_counter
    no_emojis = emoji_count(tweet)

    if no_emojis > 0:
      tweet = emoji.demojize(tweet)
      tweet = tweet.replace(":"," ")
      tweet = ' '.join(tweet.split())

    emoji_count = sum([tweet.count(emoj) for emoj in all_emojis ])
    tweet = emoji.demojize(tweet, delimiters=("", ""))
  
    return tweet

#clean the tweets in the text column (version 2)
df_modified['text'] = df_modified['text'].apply(cleanTweet2)
```

```python
#remove remaining emojis
def removeEmoji(tweet):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', tweet)

df_modified['text'] = df_modified['text'].apply(removeEmoji)
```

```python
def cleanTweet3(tweet):

  #fix misspelled words
  tweet = TextBlob(tweet).correct()

  #replace contractions with full forms
  #creating an empty list
  expanded_words = []
  for word in tweet.split():
    #using contractions.fix to expand the shotened words
    expanded_words.append(contractions.fix(word))
    tweet = ' '.join(expanded_words)

  return tweet

#clean the tweets in the text column (version 3)
df_modified['text'] = df_modified['text'].apply(cleanTweet3)
```
